```python
import pandas as pd
df_nozzleresults = pd.read_csv("xy_area_stacked_all.csv")
df_nozzleresults.name = "xy_area_stacked_all.csv"

df_baseresults = pd.read_csv("test_log_base_width_mar7.csv")
df_baseresults[['Base_Wafer_ID','Die_N']] = df_baseresults['Wafer_DieNo'].str.split('-',expand=True).apply(pd.to_numeric)
df_baseresults.name = "test_log_base_width_mar7.csv"

df_baseresults2 = pd.read_csv("test_log_base_wafer_volume_mar7.csv")
df_baseresults2[['Base_Wafer_ID','Die_N']] = df_baseresults2['Module_Nbr'].str.split('-',expand=True).apply(pd.to_numeric)
df_baseresults2.name = "test_log_base_wafer_volume_mar7.csv"
```

```python
df_wafer_map = pd.read_csv("Base_Nozzle_Wafer_Mapping.csv")
df_wafer_map.name = "Base_Nozzle_Wafer_Mapping.csv"
df_wafer_map[ 'Nozzle_Wafer_ID' ] = df_wafer_map['consumable_unit_id'].apply(pd.to_numeric)

df_wafer_map2 = pd.read_csv("Base_Nozzle_Wafer_Mapping0.csv")
df_wafer_map2.name = "Base_Nozzle_Wafer_Mapping0.csv"
df_wafer_map2[ 'Nozzle_Wafer_ID' ] = df_wafer_map2['consumable_unit_id'].apply(pd.to_numeric)
df_wafer_map2[ 'nozzle_product' ] = df_wafer_map2['nozzleproduct'].apply(pd.to_numeric)

df_wafer_map3 = pd.read_csv("Base_Nozzle_Wafer_Mapping1.csv")
df_wafer_map3.name = "Base_Nozzle_Wafer_Mapping1.csv"
df_wafer_map3[ 'Nozzle_Wafer_ID' ] = df_wafer_map3['consumable_unit_id'].apply(pd.to_numeric)
df_wafer_map3[ 'nozzle_product' ] = df_wafer_map3['nozzleproduct'].apply(pd.to_numeric)


df_wafer_map11 = pd.read_csv("nozzle2.csv")
df_wafer_map11.name = "nozzle2.csv"
df_wafer_map11[ 'Nozzle_Wafer_ID' ] = df_wafer_map11['consumable_unit_id'].apply(pd.to_numeric)
df_wafer_map11[ 'nozzle_product' ] = df_wafer_map11['nozzleproduct'].apply(pd.to_numeric)

df_wafer_map12 = pd.read_csv("nozzle_1.csv")
df_wafer_map12.name = "nozzle_1.csv"
df_wafer_map12[ 'Nozzle_Wafer_ID' ] = df_wafer_map12['consumable_unit_id'].apply(pd.to_numeric)
df_wafer_map12[ 'nozzle_product' ] = df_wafer_map12['nozzleproduct'].apply(pd.to_numeric)

df_wafer_map13 = pd.read_csv("nozzle data11.csv")
df_wafer_map13.name = "nozzle data11.csv"
df_wafer_map13[ 'Nozzle_Wafer_ID' ] = df_wafer_map13['consumable_unit_id'].apply(pd.to_numeric)
df_wafer_map13[ 'nozzle_product' ] = df_wafer_map13['nozzleproduct'].apply(pd.to_numeric)

df_wafer_map4 = pd.concat([df_wafer_map, df_wafer_map2, df_wafer_map3, df_wafer_map11, df_wafer_map12, df_wafer_map13], sort=False)
df_wafer_map4.head(5)
```
